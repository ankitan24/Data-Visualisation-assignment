QUES 4 -Generate a dataset with categories and numerical values. Visualize the distribution of a numerical variable across different categories

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import pandas as pd
import numpy as np

# Set up the output to display in a notebook (if you're using Jupyter Notebook)
output_notebook()

# Generate a dataset
np.random.seed(42)  # For reproducibility
categories = ['Category A', 'Category B', 'Category C', 'Category D']
data = {
    'Category': np.random.choice(categories, size=200),  # Randomly assign categories
    'Value': np.random.randn(200) * 10 + 50  # Generate numerical values
}

# Create a DataFrame
df = pd.DataFrame(data)

# Create a Bokeh figure for box plot
p = figure(title="Distribution of Values Across Categories",
           x_axis_label='Category', y_axis_label='Value',
           x_range=categories, width=800, height=400)

# Calculate quartiles for box plot
q1 = df.groupby('Category')['Value'].quantile(0.25)
q2 = df.groupby('Category')['Value'].quantile(0.50)
q3 = df.groupby('Category')['Value'].quantile(0.75)

# Add boxes to the plot
p.quad(top=q3, bottom=q1, left=np.arange(len(categories)) - 0.2,
       right=np.arange(len(categories)) + 0.2, fill_color='green', line_color='black', alpha=0.6)

# Add median line
p.line(np.arange(len(categories)), q2, line_color='black', line_width=2)

# Show the plot
show(p)
```

QUES 5-Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set a random seed for reproducibility
np.random.seed(42)

# Number of samples
n_samples = 500

# Generate correlated data
# First, create an independent variable
x1 = np.random.normal(size=n_samples)

# Generate other variables that are correlated with x1
x2 = 0.8 * x1 + np.random.normal(scale=0.2, size=n_samples)  # Strong correlation
x3 = -0.5 * x1 + np.random.normal(scale=0.3, size=n_samples)  # Negative correlation
x4 = np.random.normal(size=n_samples)  # Uncorrelated with others

# Create a DataFrame
data = pd.DataFrame({'Feature_1': x1, 'Feature_2': x2, 'Feature_3': x3, 'Feature_4': x4})

# Calculate the correlation matrix
correlation_matrix = data.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

QUES 3 - . Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set the random seed for reproducibility
np.random.seed(20)

# Generate the sales dataset
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

# Pivot the data to create a matrix for the heatmap
heatmap_data = df.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='mean')

# Create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, fmt=".1f", cmap='coolwarm', cbar_kws={'label': 'Average Sales'})
plt.title('Heatmap of Sales Across Months and Days')
plt.xlabel('Month')
plt.ylabel('Day')
plt.xticks(rotation=45)  # Rotate month labels for better visibility
plt.yticks(rotation=0)    # Keep day labels horizontal
plt.show()
```

QUES 4-   Using the given x and y data, generate a 3D surface plot to visualize the function


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate the x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Create a DataFrame from the data
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create a 3D surface plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(x, y, z, cmap='viridis', edgecolor='PURPLE')

# Set labels and title
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
ax.set_title('3D Surface Plot of z = sin(sqrt(x^2 + y^2))')

# Show the plot
plt.show()
```

QUES 5-. Create a Bokeh heatmap using the provided dataset.

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256
import numpy as np
import pandas as pd

# Set up the output to display in a notebook (if you're using Jupyter Notebook)
output_notebook()

# Generate random heatmap data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Create a DataFrame for the heatmap
df = pd.DataFrame(data_heatmap, index=y, columns=x)

# Convert DataFrame to long format for Bokeh
heatmap_data = df.reset_index().melt(id_vars='index')
heatmap_data.columns = ['y', 'x', 'value']

# Create a Bokeh figure
p = figure(title="Heatmap",
           x_axis_label='X', y_axis_label='Y',
           x_range=(0, 1), y_range=(0, 1),
           width=600, height=600)

# Create a color mapper
mapper = linear_cmap(field_name='value', palette=Viridis256, low=df.values.min(), high=df.values.max())

# Add rectangles to represent the heatmap
p.rect(x='x', y='y', width=0.1, height=0.1, source=heatmap_data,
       fill_color=mapper, line_color=None)

# Show the plot
show(p)
```
